```python
# Importing the libraries

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

# Ignore harmless warnings 

import warnings 
warnings.filterwarnings("ignore")

# Set to display all the columns in dataset

pd.set_option("display.max_columns", None)

pd.set_option("display.max_rows", None)

# Import psql to run queries 

import pandasql as psql

data = pd.read_csv('full_data.csv',header=0)
```

# Pieplot

```python
plt.figure(figsize=(5,5))
data['hospital_death'].value_counts().plot(kind='pie')
```

# Barplot

```python
plt.figure()
data['hospital_death'].value_counts().plot(kind='bar')
```

# HeatMap

```python
plt.figure(figsize = (100, 100))
sns.heatmap(data.corr(), cbar=True, square=True, annot=True, cmap='coolwarm',vmin = -1)
```

# Catplot

Relation between age ,ethinincty and hospital_death using catplot( kind = violin)

```python
g = sns.catplot(x="hospital_death", y="age", hue='ethnicity', data=data, kind="violin", aspect = 3)
```

RELATION BETWEEN BMI AND hospital_death USING CATPLOT(KIND = strip)

```python
g = sns.catplot(x="hospital_death", y="bmi", data=data, kind="strip")
```

h1_sysbp_max and h1_sysbp_noninvasive_max has correlation = 0.98

```python
plt.scatter(data['h1_sysbp_max'],data['h1_sysbp_noninvasive_max'])
plt.show()
```

apache_3j_diagnosis and weight has correlation=0.00085

```python
plt.scatter(data['weight'],data['apache_3j_diagnosis'])
plt.show()
```

# Density Plot

```python
sns.set(style="darkgrid")
fig = sns.kdeplot(data['heart_rate_apache'], shade=True, color="r")
fig = sns.kdeplot(data['d1_diasbp_max'], shade=True, color="b")
plt.xlabel("---------------")
plt.ylabel("Density")
plt.legend(["heart_rate_apache", "d1_diasbp_max"])
plt.show()
```
